# 2. Data Cleaning & Feature Engineering

Goal: produce a clean, model-ready dataset and save it to `data/processed/`.

```python
import sys
sys.path.append('..')

import pandas as pd
import numpy as np
from src.preprocess import fill_missing, encode_categoricals, remove_outliers
from src.features import add_features, log_transform_target

train = pd.read_csv('../data/raw/train.csv')
test  = pd.read_csv('../data/raw/test.csv')

print('Train shape:', train.shape)
print('Test shape: ', test.shape)
```

## Remove outliers & separate target

```python
train = remove_outliers(train)
print('After outlier removal:', train.shape)

y_train = log_transform_target(train['SalePrice'])
train = train.drop(columns=['SalePrice'])
```

## Combine train + test for consistent encoding

```python
n_train = len(train)
combined = pd.concat([train, test], axis=0).reset_index(drop=True)
print('Combined shape:', combined.shape)
```

## Preprocessing pipeline

```python
combined = combined.drop(columns=['Id'], errors='ignore')
combined = fill_missing(combined)
combined = add_features(combined)
combined = encode_categoricals(combined)

print('After preprocessing:', combined.shape)
```

## Split back and save

```python
X_train = combined.iloc[:n_train]
X_test  = combined.iloc[n_train:]

X_train.to_csv('../data/processed/X_train.csv', index=False)
X_test.to_csv('../data/processed/X_test.csv', index=False)
y_train.to_csv('../data/processed/y_train.csv', index=False)

print('Saved to data/processed/')
print('X_train:', X_train.shape)
print('X_test: ', X_test.shape)
print('y_train:', y_train.shape)
```
